Elimdeki tüm textlerden unique oan kelimeleri çekiyorum.(STEM VE LEMİTAZİNG YAPTIKTAN SONRA) bu bizim sözlükümüz oluyor.

Sonra her text için bu sözlüge göre hangi kelime ne kadar kullanıldı bir vektör oluşturuyorsun.Burdan en fazla kullanılan kelimelerden bir Inferance ile beraber sonucu almış oluyoruz.

Bu BOW yöntemin biraz daha fazla vektörü uzun olabilir bu dezavantaj olabilir.

Sparse Representation: Vektörün çogunun 0 olması demektir.

Kelimelerin anlamlarını tam olarak yakalayamayabilir.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(f'../spam.csv')
df.head()
```

```python
df.Category.value_counts()
```

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.2, stratify=df.spam)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()

X_train_cv = v.fit_transform(X_train.values)
X_train_cv
```

```python
X_train_cv.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

```python
x_test_cv = v.transform(X_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(x_test_cv)

print(classification_report(y_test, y_pred))
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]

emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```
